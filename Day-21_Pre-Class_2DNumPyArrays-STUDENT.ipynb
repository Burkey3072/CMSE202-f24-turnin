## Part 1: 2D NumPy arrays

In this assignment, we are going to spend some time thinking about higher dimensional arrays. In using the term "dimension" when talking about arrays, it can _almost_ help to think about these like geometrical shapes or objects. A line only has a length to it, so we call it a one dimensional object. A square, however, has a length _and_ a width, so we call it a two dimensional object. Though this metaphor can start to break down as the shapes become more complex, it works for our 1D and 2D cases. A 1D array is simply a single array of value, like `array([1,2,3,4,5])`. A 2D array, which we are going to introduce here today, is as bit more complex. 2D arrays are very commonly used in computational modeling and data analysis for reasons that will hopefully be clear after today.

First, let's make a 2D NumPy array to represent the _Battleship_ grid.  We will initialize all of the values to **0** using the `np.zeros` function in NumPy.

```python
##Import numpy and alias it first
import numpy as np
##Make a board of zeros of shape 10 x 10
board = np.zeros((10, 10))
board
```

**Now that we have our board made**, we can use 2D indices to try and recreate the following _Battleship_ board. 

For squares where a ship is present, we will have the corresponding element of the array have a value of **1**. We don't want to define _each individual element_ of the array. That would take a long time. But, we can use 2D indexing to speed things up a bit and change ranges of elements at a time. For example, the ship on the right-hand side of the board which occupies three squares could be placed on our board using the following line of code: `board[3:6,8] = 1`, which __indexes__ the rows 3, 4, and 5 (not 6 because the endpoint is not included!) of the board and column 8 of the board, then sets the __values__ at those locations to be 1.

Let's start placing ships!

Once you have placed the ships, look at the entries of 2D array and then use the `matplotlib` function `imshow` to visualize the board and make sure that it matches our image. What do you notice about the values in the 2D array and how do they correspond to the image?

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Place first boat
board[3:6,8] = 1
# Place second boat
board[0, 1:4] = 1

# Print the 2D Array
print(board)

# Display the board
plt.imshow(board)
plt.axis('off') # Don't include the axes!
```

#### Using Masking for Filtering Data

Let's use masking to try and answer the question we posed above. **Make sure you read the code comments in the following cells to understand what the code is doing.** This is all very similar to what was explained in the video on masking. 

```python
##First, let's see what our masking condition returns. The horsepower is the column at index 1
##Remember, we only want to see whether the horsepower column of our dataset is greater than 300. If we
##don't specify that when we write our masking condition, then it will check every value in our array.
##Another way of saying this is that we will be masking every value in the array that's greater than 300,
##not JUST the horsepower like we want
mask = car_data[:,1] >= 300
```

Images are essentially 3D arrays of RGB (Red, Green, Blue) vectors. Each pixel in the image is represented by an array of three values, corresponding to the intensities of red, green, and blue in that pixel. These values range from 0 to 255.

* **White pixels** are represented by the array [255, 255, 255], indicating maximum intensity of red, blue, and green.
* **Black pixels** are represented by the array [0, 0, 0], indicating no color or zero intensity.

Here's an illustrative image to help you visualize the array structure of an image:


<img src="https://i.ibb.co/cYRkr2x/image-structure.png" alt="image-structure" border="0" width=800>

Image Credit: Sarah Tymochko

When we load an image into Python, it recognizes the structure of the file and loads it as a 3D array, with each pixel represented by an RGB vector. However, for image analysis, we might want to work with each color channel separately. This allows us to analyze the red, green, and blue components of the image individually.

The following code separates out the red, green, and blue components for each pixel in our image:



```python
##The following code separates out the red, blue, and green components for each of the pixels in our image
red_array, green_array, blue_array = image[:,:,0], image[:,:,1], image[:,:,2]
```



In this code, `image[:,:,0]` extracts the red component, `image[:,:,1]` extracts the green component, and `image[:,:,2]` extracts the blue component. The result is three 2D arrays, each representing one color channel of the image.

```python
# Let's plot each channel and see what it looks like
fig, axs = plt.subplots(1, 3, figsize=(12, 3))

# Plot red array
red_plot = axs[0].imshow(red_array, cmap='Reds')
axs[0].set_title('Red Array')
fig.colorbar(red_plot, ax=axs[0])  # Add colorbar to the plot

# Plot green array
green_plot = axs[1].imshow(green_array, cmap='Greens')
axs[1].set_title('Green Array')
fig.colorbar(green_plot, ax=axs[1])  # Add colorbar to the plot

# Plot blue array
blue_plot = axs[2].imshow(blue_array, cmap='Blues')
axs[2].set_title('Blue Array')
fig.colorbar(blue_plot, ax=axs[2])  # Add colorbar to the plot

fig.tight_layout()
```

Let's break down this code:

1. `fig, axs = plt.subplots(1, 3, figsize=(12, 3))`: This line creates a figure object (`fig`) and an array of axes objects (`axs`). The `subplots` function is called with parameters `(1, 3)`, which means that the figure will contain 1 row and 3 columns of subplots. The `figsize` parameter sets the width and height of the figure in inches.

2. `red_plot = axs[0].imshow(red_array, cmap='Reds')`: This line creates a 2D image of the `red_array` on the first subplot (indexed by 0). The `cmap` parameter is set to `'Reds'`, which means that the data values in `red_array` will be mapped to shades of red.

3. `axs[0].set_title('Red Array')`: This line sets the title of the first subplot to 'Red Array'.

4. `fig.colorbar(red_plot, ax=axs[0])`: This line adds a colorbar to the first subplot. The colorbar is associated with the image `red_plot`, and it provides a visual representation of the mapping between data values in `red_array` and shades of red.

5. The next few lines do the same for the `green_array` and `blue_array`, but map the data values to shades of green and blue, respectively.

6. `fig.tight_layout()`: This line adjusts the positions of the subplots so that there is no overlap between them.

So, in this code, three 2D images are created from `red_array`, `green_array`, and `blue_array`, respectively. Each image is displayed in a separate subplot with a colorbar that shows the mapping between data values and colors. The colors used are shades of red, green, and blue, respectively. The `tight_layout` function is called to ensure that the subplots do not overlap.
